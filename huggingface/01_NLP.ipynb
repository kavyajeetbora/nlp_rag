## Summarization task

```python
summarizer = pipeline(task="summarization",
                      model="facebook/bart-large-cnn",
                      torch_dtype = torch.bfloat16
            )
```

```python
story = '''
Sunil Kumar Gupta (Zahan Kapoor) is a rookie officer in the notoriously difficult Tihar Jail in Delhi, aiming to approach his new role with honor.
But soon he finds that there is corruption across the prison complex, as prison guards work with the various inmate gangs to
smuggle drugs and alcohol in and out for a profit.
The factions, as told to Gupta, are: the Tyagi Gang, which handles the lucrative drug trade; the Rana gang, who coordinates alcohol
trade and are eyeing a collaboration with the Tyagis; and the Sardar group, much less a gang and rather a group of Sikhis that chooses to keep to themselves
'''
```

```python
summary = summarizer(story)
```

## Sentence Transformers

- Helps to capture semantics of a given text in numerical format also called embeddings
- We can use this for information retrieval, clustering or grouping

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentences1 = ["This is an example sentence", "Each sentence is converted"]
sentences2 = ["It is one example sentence", "Each other sentence is converted"]

embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2)

print(embeddings1.shape, embeddings2.shape)
```

Find the similarities

```python
from sentence_transformers import util

cosine_scores = util.cos_sim(embeddings1, embeddings2)

for i in range(len(sentences1)):
    print()
```
